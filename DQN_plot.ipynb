```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
def rolling_average(data, *, window_size):
    """Smoothen the 1-d data array using a rollin average.

    Args:
        data: 1-d numpy.array
        window_size: size of the smoothing window

    Returns:
        smooth_data: a 1-d numpy.array with the same size as data
    """
    assert data.ndim == 1
    kernel = np.ones(window_size)
    smooth_data = np.convolve(data, kernel) / np.convolve(
        np.ones_like(data), kernel
    )
    return smooth_data[: -window_size + 1]
```

```python
dqn_data = pd.read_csv('cart_dqn.csv')
double_data = pd.read_csv('cart_double_dqn.csv')

dqn_return = dqn_data.returns
dqn_lengths = dqn_data.lengths

double_returns = double_data.returns
double_lengths = double_data.lengths

plt.plot(rolling_average(dqn_return, window_size= 200))
plt.plot(rolling_average(double_returns, window_size= 200))
plt.legend(['DQN','Double DQN '])
plt.title('Cartpole - Returns')
plt.grid()

plt.show()

plt.plot(rolling_average(dqn_lengths, window_size= 200))
plt.plot(rolling_average(double_lengths, window_size= 200))
plt.legend(['DQN','Double DQN '])
plt.title('Cartpole - Lengths')
plt.grid()

plt.show()
```
