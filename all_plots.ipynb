```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
def rolling_average(data, *, window_size):
    """Smoothen the 1-d data array using a rollin average.

    Args:
        data: 1-d numpy.array
        window_size: size of the smoothing window

    Returns:
        smooth_data: a 1-d numpy.array with the same size as data
    """
    assert data.ndim == 1
    kernel = np.ones(window_size)
    smooth_data = np.convolve(data, kernel) / np.convolve(
        np.ones_like(data), kernel
    )
    return smooth_data[: -window_size + 1]
```

# DQN v/s Dueling v/s Double


```python
#cartpole
min_range = (min( len(cart_dqn_returns), len(cart_dueling_returns), len(cart_double_returns) ))

plt.plot(rolling_average(cart_dqn_returns[0:min_range],window_size = 100))
plt.plot(rolling_average(cart_dueling_returns[0:min_range],window_size = 100))
plt.plot(rolling_average(cart_double_returns[0:min_range],window_size = 100))

plt.xlabel('Episodes')
plt.ylabel('Return')

plt.title('Cartpole Environment')
plt.legend(['DQN','Dueling DQN', 'Double DQN'])
```
